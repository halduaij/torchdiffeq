```python
# test_radau_dirk.py
# -----------------------------------------------------------
import time
import torch
from torchdiffeq._impl.radau import RadauSolver      # wrapper you re‑added
from torchdiffeq import odeint_adjoint               # could also use odeint

device = "cuda" if torch.cuda.is_available() else "cpu"

# 1. stiff Van‑der‑Pol oscillator ------------------------------------------------
class VDP(torch.nn.Module):
    def __init__(self, mu_val=1000.0):
        super().__init__()
        # treat μ as a learnable parameter so we can test gradients
        self.mu = torch.nn.Parameter(torch.tensor(mu_val, dtype=torch.float64))

    def forward(self, t, y):
        y1, y2 = y[..., 0], y[..., 1]
        mu = self.mu
        dydt = torch.stack([
            y2,
            mu * (1 - y1**2) * y2 - y1
        ], dim=-1)
        return dydt

# 2. initial condition & time grid ---------------------------------------------
y0 = torch.tensor([2.0, 0.0], dtype=torch.float64, device=device)
t  = torch.linspace(0.0, 0.1, 51, dtype=torch.float64, device=device)  # 50 steps

model = VDP().to(device)

# 3. forward integration with your Radau DIRK solver ---------------------------
start = time.time()
y_traj = odeint_adjoint(
    model, y0, t,
    method='radau',           # <-- uses your patched implementation
    rtol=1e-2, atol=1e-2
)
elapsed = time.time() - start

# 4. simple loss = L2 of final state, back‑prop -------------------------------
loss = (y_traj[-1]**2).sum()      # scalar
loss.backward()

# 5. diagnostics ---------------------------------------------------------------
print("Integration finished in %.3f s" % elapsed)
print("y(t=0.1) =", y_traj[-1].detach().cpu().numpy())
print("grad dLoss/dμ =", model.mu.grad.item())
```

```python
# stiff_benchmark.py  — final working version
import time, torch, pandas as pd, matplotlib.pyplot as plt
from torchdiffeq import odeint
from torchdiffeq._impl.radau import RadauSolver     # your patched class

torch.set_default_dtype(torch.float64)
device = "cpu"

class VDP(torch.nn.Module):
    def __init__(self, mu): super().__init__(); self.mu = mu
    def forward(self, t, y):
        y1, y2 = y[..., 0], y[..., 1]
        return torch.stack([y2, self.mu*(1-y1**2)*y2 - y1], dim=-1)

def runtime(method, mu):
    t  = torch.linspace(0., 1.0, 250, device=device)
    y0 = torch.tensor([2., 0.], device=device)
    tic = time.perf_counter()
    odeint(VDP(mu), y0, t, method=method, rtol=1e-6, atol=1e-6)
    return time.perf_counter() - tic

mus  = [1, 10, 50, 100, 400, 800, 1600, 3200]
rows = [{"mu": mu,
         "Hybrid Radau": runtime(RadauSolver, mu),
         "Dopri5":        runtime("dopri5",   mu)}
        for mu in mus]

df = pd.DataFrame(rows)
print(df.to_string(index=False))

plt.figure(figsize=(6,4))
plt.plot(df["mu"], df["Hybrid Radau"], "o-", label="hybrid Radau")
plt.plot(df["mu"], df["Dopri5"],      "o-", label="explicit Dopri5")
plt.xscale("log"); plt.xlabel("μ  (stiffness)"); plt.ylabel("runtime (s)")
plt.title("Runtime vs stiffness"); plt.legend(); plt.tight_layout(); plt.show()
```

```python
import time, torch, pandas as pd, matplotlib.pyplot as plt
from torchdiffeq._impl.radau import RadauSolver   # your patched solver

torch.set_default_dtype(torch.float64)
device = "cpu"            # flip to "cuda" for GPU timing

# ---------------------------------------------------------------------------
class VDP(torch.nn.Module):
    def __init__(self, mu): super().__init__(); self.mu = mu
    def forward(self, t, y):
        y1, y2 = y[..., 0], y[..., 1]
        return torch.stack([y2, self.mu*(1-y1**2)*y2 - y1], dim=-1)

def make_rhs(mu):
    base = VDP(mu).to(device)
    def rhs(t, y, perturb=None): return base(t, y)     # accept *perturb*
    rhs.callback_step = rhs.callback_accept_step = rhs.callback_reject_step = lambda *a,**k: None
    return rhs

def runtime_radau(mu, t_end=1.0, N=250):
    t  = torch.linspace(0., t_end, N, device=device)
    y0 = torch.tensor([2., 0.], device=device)
    rhs = make_rhs(mu)
    solver = RadauSolver(func=rhs, y0=y0,
                         rtol=1e-3, atol=1e-3,
                         norm=lambda z: z.abs().max())
    tic = time.perf_counter()
    solver.integrate(t)
    return time.perf_counter() - tic

# ---------------------------------------------------------------------------
mus   = [1, 10, 50, 100, 400, 800, 1600, 3200]
times = [runtime_radau(mu) for mu in mus]

df = pd.DataFrame({"μ": mus, "Hybrid Radau runtime (s)": times})
print(df.to_string(index=False))

plt.figure(figsize=(6,4))
plt.plot(mus, times, "o-")
plt.xscale("log")
plt.xlabel("μ  (stiffness parameter, log scale)")
plt.ylabel("runtime  (s)")
plt.title("Hybrid Radau runtime vs stiffness")
plt.tight_layout()
plt.show()
```
